## **Valor promedio por tipo de propiedad**

```python
df_price_type = df.groupby('Tipo')[['Valor']].mean().sort_values(by='Valor')
df_price_type
```

```python
df_price_type.plot(kind='barh', figsize=(12,8))
```

```python
df.Tipo.unique()
```

```python
comerciales = ['Conjunto Comercial/Sala', 'Edificio Completo',
       'Tienda/Salón', 'Casa Comercial',
       'Terreno Estándar', 'Cochera/Estacionamiento',
       'Galpón/Depósito/Almacén', 'Tienda en Centro Comercial',
       'Hotel', 'Loteo/Condominio','Industria']
```

```python
df_residenciales = df.query('@comerciales not in Tipo')
```

```python
df_price_type = df_residenciales.groupby('Tipo')[['Valor']].mean().sort_values(by='Valor')
df_price_type
df_price_type.plot(kind='barh', figsize=(12,8))
```
